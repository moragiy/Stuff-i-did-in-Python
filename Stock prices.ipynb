```python
#create the reference times

today = date.today()
last_monday = today + relativedelta(weekday=MO(-1))
last_last_monday= today + relativedelta(weekday=MO(-2))
monday_three_month_ago=today + relativedelta(weekday=MO(-13))
monday_52_weeks_ago=today + relativedelta(weekday=MO(-53))

print (today)
print (last_monday)
print (last_last_monday)
print (monday_three_month_ago)
print (monday_52_weeks_ago)
```

# sheet 1

```python
#LIST OF STOCKS ENDING: https://help.yahoo.com/kb/SLN2310.html

#Interval required 5 minutes
Stocks = ["ASELS.IS", "BA", "BA.L", "CAE", "RTX", "AIR.PA", "ERJ" , "ESLT", "LDO.MI", "TDY", "GD", "GNTX", "GRMN", "HO.PA","HON",
          "LHX", "ISSC", "KOG.OL", "LMT", "NOC", "RHM.F","SAAB-B.ST","%5EDJI", "%5EGSPC", "^IXIC" ,"047810.KS"]
ls = []
for i in Stocks:
    df = yf.download(tickers=i, start=monday_52_weeks_ago,end=today, interval='1wk')
    df["ticker"] = i
    ls.append(df)
    #print(df)
```

```python
#Stocks
```

```python
print(df)
```

```python
df = pd.concat(ls)
```

```python
df = df.reset_index()
```

```python
df["ticker+Date"] = df['ticker']+' '+df['Date'].apply(lambda x: str(x)[:10])
```

```python
df= df[["ticker+Date",'ticker','Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
df
```

```python
df1 = df[['ticker','Date','Close']]
df1['Date'] = df1['Date'].apply(lambda x: str(x)[:10])
df1.set_index('Date', inplace=True)
df1 = pd.pivot_table(df1, values = 'Close', index = ['ticker'], columns= ['Date'])
df1
```
